```python
# Calculate average price by house condition and sort in descending order
avg_price = data.groupby("condition of the house")["Price"].mean().sort_values(ascending=False)

# Plot the bar chart
plt.figure(figsize=(8, 5))
avg_price.plot(kind='bar', color='skyblue')

# Add labels and title
plt.xlabel("Condition of the House", fontsize=12)
plt.ylabel("Average Price (INR)", fontsize=12)
plt.title("Average House Price by Condition", fontsize=14)

# Customize ticks
plt.xticks(rotation=0)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Annotate each bar with its value
for index, value in enumerate(avg_price):
    plt.text(index, value + 10000, f"{int(value):,}", ha='center', va='bottom', fontsize=9)

# Show the plot
plt.tight_layout()
plt.show()
```
